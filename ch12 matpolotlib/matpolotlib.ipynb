# Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
# 화면에 그래프 그리기
%matplotlib inline
# Korean Setting , Minus 표시
mpl.rcParams['axes.unicode_minus'] = False
# family = 폰트 이름 size = 사이즈 크기
mpl.rc('font', family='D2Coding', size=25)
# Graph 크기 변경
plt.rcParams["figure.figsize"]= (15,9)
```

```python
# ex 12-1
data = [20,30,40]
x = [1,2,3]
plt.plot(x, data)
plt.show()
```

```python
# ex 12-2
# plot = line graph, bar = bar graph
plt.plot(['Kim','Lee', 'Kang'], [85,88,90], color='red', marker='o')
# title 
plt.title('English Score of three students')
# x label name
plt.xlabel('Student Name ')
# y label name
plt.ylabel('Score')
plt.show()
```

```python
# 12-3
xdata = ['안지영', '홍지수', '황예린']
ydata = [90,85,88]
plt.plot(xdata, ydata)

plt.title('세 학생의 국어 성적')
plt.xlabel('이름')
plt.ylabel('성적')

plt.show()
```

```python
# 12-4
xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]
plt.plot(xdata, ydata1, label='국어')
plt.plot(xdata, ydata2, label='영어')
# legend = 범례 upper center -> 중앙 상단
plt.legend(loc='upper center')

plt.title('세 학생의 국어, 영어 성적')
plt.show()
```

```python
# 12-5
xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]
ydata3 = [85,97,78]
ydata4 = [92,88,82]

# plt.plot(xdata, ydata1, label='국어', color='r', linestyle='-', marker ='o')
# plt.plot(xdata, ydata2, label='영어', color='b', linestyle='--', marker ='x')
# plt.plot(xdata, ydata3, label='수학', color='m', linestyle='-.', marker ='s')
# plt.plot(xdata, ydata4, label='사회', color='c', linestyle=':', marker ='d')

# color linestyle marker 간소화
# linewidth = 라인 굵기, markersize = 마커 굵기
plt.plot(xdata, ydata1, 'ro-' , label='국어', linewidth=3, markersize=10)
plt.plot(xdata, ydata2, 'bx--', label='영어', linewidth=3, markersize=10)
plt.plot(xdata, ydata3, 'ms-', label='수학', linewidth=3, markersize=10)
plt.plot(xdata, ydata4, 'c:d', label='사회', linewidth=3, markersize=10)


plt.title('세 학생의 네 과목 성적')
# 최적의 장소에 범례를 알아서 설정해라
plt.legend(loc='best')
plt.show()
```

```python
# 12-6
xdata= list(range(30))
ydata = []

for x in xdata:
    y = 2*x
    ydata.append(y)

plt.plot(xdata, ydata, 'm',label='y=2x', linewidth=5)
# x 축 범위 설정
plt.xlim(0, 31)
plt.xticks(list(range(0,31,2)))

# y 축 범위 설정
plt.ylim(0, 65)
plt.yticks(list(range(0,66,5)))

# show grid
plt.grid(True)
plt.show()
```

```python
# sin graph
x_data = np.linspace(-np.pi, np.pi, 201)

plt.title('Sin')
plt.plot(x_data, np.sin(x_data), 'c',linewidth = 3)



plt.grid(True)
plt.show()
```

```python
# cos graph

x_data = np.linspace(-np.pi, np.pi, 201)

plt.title('cos')
plt.plot(x_data, np.cos(x_data), 'g',linewidth = 3)
plt.grid(True)
plt.show()
```

```python
x_data = np.linspace(0, 2*np.pi, 201)
y_data1 = np.sin(x_data)
y_data2 = np.cos(x_data)

plt.title('Sin, Cos')
plt.plot(x_data, y_data1, 'r',label = 'Sin', linewidth = 3)
plt.plot(x_data, y_data2, 'b',label = 'Cos', linewidth = 3)
plt.plot(3,0,'k')
plt.legend(loc='best')
plt.show()
```

```python
x_data = np.arange(0, 2*np.pi+0.01, 0.01)
y_data1 = np.sin(x_data)
y_data2 = np.cos(x_data)

plt.title('Sin, Cos')
plt.plot(x_data, y_data1, 'r',label = 'Sin', linewidth = 3)
plt.plot(x_data, y_data2, 'b',label = 'Cos', linewidth = 3)

# X 축 선 그리기
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axhline(y=1, color='k', linewidth=0.5)
plt.axhline(y=-1, color='k', linewidth=0.5)

# Y 축 선 그리기
plt.axvline(x=0, color='k', linewidth=0.5)
plt.axvline(x=np.pi, color='k', linewidth=0.5)
plt.axvline(x=2*np.pi, color='k', linewidth=0.5)



plt.legend(loc='best')
plt.show()
```

```python
#-3 ~ 3 sigmoid
x_data = np.linspace(-3,3,10000)
# x_data = np.arange(-3,3+0.1, 0.000001)
y_data = 1/(1 + np.exp(-x_data))

plt.plot(x_data, y_data, 'r',label = 'Sigmoid', linewidth = 3)

# plt.axhline(y=1, color='k', linewidth=0.5)
# plt.axhline(y=0.5, color='k', linewidth=0.5)
plt.axhline(y=0, color='k', linewidth=3)

# plt.axvline(x=-3, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=3)
# plt.axvline(x=3, color='k', linewidth=0.5)
plt.grid(True)
plt.show()
```
